```python
import matplotlib.pyplot as plt
```

```python
def isPrime(n):
    for i in range(2, n):
        if i*i > n:
            return True
        if n % i == 0:
            return False
```

```python
f = lambda n: int("0b" + (str(bin(n))[-1:1:-1]),2)
M = 13
N = 2**M
x = [i for i in range(N) if isPrime(i)]
x = [i for i in range(N) if i % 2 != 0 and i % 3 != 0] # Thisis all that is required to get parallelograms
y = [n - f(n) for n in x]
```

```python
fig, ax = plt.subplots(figsize=(8,8))
ax.plot(list(x),list(y),'r.')
ax.set_aspect(aspect=1)
a = ax.set_xticks([2**i for i in range(1,M+1)])
a = ax.set_xticklabels(range(1,M+1))
a = ax.set_yticks([-2**i for i in range(1,M)] + [0] + [2**i for i in range(0,M)])
a = ax.set_yticklabels([-i for i in range(1,M)] + [0] + [i for i in range(0,M)])
```

```python
def gcd(n,m):
    k = n % m
    if k == 0:
        return m
    else:
        return gcd(m, k)
```

```python
N = 1000
M = 1000
# Rationals of the form i/j with 0 <= i < 40, 0 < j < 40
G = [(i/j, i + j) for i in range(N) for j in range(1, M) if gcd(i,j) == 1]
G.sort(key=lambda a: a[0])        
```

The graph of $f(m/n)=m+n$ if $\gcd(m,n)=1$ is pretty complicated.

```python
xs = [x for x, y in G]
ys = [y for x, y in G]
plt.plot(xs,ys,'r.')
```

To consider whether this is continuous pick a target point $a$ and some small $\epsilon>0$. Does it look like $f(x)$ is close to $f(a)$ for $x$ close to $a$?

```python
epsilon = .1
a = 5
xs = [x for x, y in G if a - epsilon < x and x < a + epsilon]
ys = [y for x, y in G if a - epsilon < x and x < a + epsilon]
```

```python
plt.plot(xs,ys,'r.')
```
